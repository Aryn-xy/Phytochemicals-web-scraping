```r
install.packages("rvest")
install.packages("dplyr")
install.packages("readxl")

library(rvest)
library(dplyr)
library(readxl)
```

```r
plant <- "Rosmarinus officinalis"
cat("Processing:", plant, "\n")

formatted_plant <- gsub(" ", "%20", plant)

# Construct the full URL for the plant's phytochemicals
plant_url <- paste0("https://cb.imsc.res.in/imppat/phytochemical/", formatted_plant)

# Try to scrape the page
page <- tryCatch({
  read_html(plant_url)
}, error = function(e) {
  cat("Error scraping", plant, "\n")
  return(NULL)
})

# Extract the table if the page loads successfully
if (!is.null(page)) {
  phytochem_table <- page %>% html_node("table") %>% html_table(fill = TRUE)

  if (!is.null(phytochem_table)) {
    phytochem_table$PlantName <- plant

    write.csv(phytochem_table, "Rosmarinus officinalis_phytochemicals.csv", row.names = FALSE)

    print(head(phytochem_table))
    View(phytochem_table)
  } else {
    cat("No phytochemicals found for", plant, "\n")
  }
}
```
